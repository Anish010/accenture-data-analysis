```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import seaborn as sns

df = pd.read_csv("final.csv")
df
```

```python
palette_color = sns.color_palette('bright')
explode = (0.1, 0, 0, 0,0) 
plt.pie(data=df,explode=explode,x= "Score", labels="Content Category", colors=palette_color, autopct='%.2f',shadow=True, startangle=10)
plt.title("Popularity percentage share from top 5 categories")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 8))
fx = sns.barplot(data=df,y="Score",x="Content Category",ax=ax)
plt.title("Top 5 categories by 'aggregate' score ")
for i in fx.containers:
    fx.bar_label(i,)
```

```python
df = df.sort_values(by=['Category'], ascending=True)
fig, ax = plt.subplots(figsize=(10, 5))
fx = sns.countplot(data=df, x="Category",ax=ax,order=df.Category.value_counts().iloc[:5].index)
for i in fx.containers:
    fx.bar_label(i,)
plt.title("Reactions count to Top 5 Category")
```
